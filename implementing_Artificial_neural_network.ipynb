```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
from google.colab import files


uploaded = files.upload()
```

```python

import io

dataset = pd.read_csv(io.BytesIO(uploaded['Churn_Modelling.csv']))
x = dataset.iloc[:,3:13]
y = dataset.iloc[:,13]
```

```python
#create dummy veriable.
geography = pd.get_dummies(x["Geography"],drop_first=True)
gender = pd.get_dummies(x["Gender"],drop_first=True)
```

```python
x = pd.concat([x,gender,geography],axis=1)
```

```python
x
```

```python
#drop unnecessory columns
x = x.drop(['Geography','Gender'],axis=1)
```

```python
x
```

```python
#spliting the dataset into train test split
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
# now lets make ANN
# import keras library and package
import keras
from keras.models import Sequential
from keras.layers import Dense
#from keras.layers import LeakeyReLU,PReLU,ELU
from keras.layers import Dropout
```

```python
#initialising ANN
classifier = Sequential()
# this is the empty neural network
```

```python
# Adding the input layer and 1st hidden layer
classifier.add(Dense(units = 6,kernel_initializer='he_uniform',activation='relu',input_dim=11))
# unit means we need 6 neurons,
# kernel_initializer use for weight initialiser when we use relu activation function he_uniform or he_normal works well.
# input dim are 11 beacause in x_train we have 11 feature
```

```python
# Adding the input layer and 2nd hidden layer
classifier.add(Dense(units = 6,kernel_initializer='he_uniform',activation='relu'))
```

```python
# Adding output layer
classifier.add(Dense(units = 1,kernel_initializer='glorot_uniform',activation='sigmoid'))
```

```python
classifier.summary()
```

```python
#compiling the ann
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
#fitting the ANN to training set
model_history = classifier.fit(x_train,y_train,validation_split=0.33,batch_size=10,epochs=100)
```

```python
# predict the test set result.
y_predict=classifier.predict(x_test)
y_predict=(y_predict>0.5)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_predict)
```

```python
#calculate the accuracy
from sklearn.metrics import accuracy_score
score =accuracy_score(y_predict,y_test)
```
